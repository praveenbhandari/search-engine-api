```python
import pandas as pd
import ast

# Function to safely parse the string representation of a list
def safe_literal_eval(s):
    try:
        return ast.literal_eval(s)
    except:
        return []  # Return an empty list in case of error

# Load the data
data = pd.read_csv("/Users/praveen/Downloads/grouped_by_case_name_with_parsed_vectors.csv")
vector_ids_column = "Parsed Vector IDs"
case_name_column = "Case Name"

# Apply the safe parsing function to convert string representations to actual lists
data['Actual Vector IDs'] = data[vector_ids_column].apply(safe_literal_eval)

# Calculate the length of each vector list
data['Vector Length'] = data['Actual Vector IDs'].apply(len)

# Create a new DataFrame with the required columns
new_data = data[[case_name_column, 'Actual Vector IDs', 'Vector Length']]

# Rename columns to match your requirements
new_data.columns = ['Case Name', 'Vector ID', 'Length']

# Print the new DataFrame to check the first few rows
print(new_data.head())

# Save the new DataFrame to a CSV file
output_file_path = "/Users/praveen/Downloads/processed_case_data_with_vector_lengths.csv"
new_data.to_csv(output_file_path, index=False)
print(f"Data successfully saved to {output_file_path}.")
```

```python
import pandas as pd
import ast

# Load the data from CSV file
data = pd.read_csv("/Users/praveen/Downloads/dev_reference (3).csv")

# Function to safely parse the string representation of a list into an actual list
def parse_vector_ids(vector_id_str):
    try:
        return ast.literal_eval(vector_id_str)
    except ValueError:
        return []  # Return an empty list in case of error

# Apply the function to each entry in the 'Vector ID' column
data['Parsed Vector IDs'] = data['Vector IDs'].apply(parse_vector_ids)

# Grouping the data by 'Case Name' and aggregating 'Parsed Vector IDs' into lists
grouped_data = data.groupby('Case Name')['Parsed Vector IDs'].agg(sum).reset_index()

# Optionally, save this grouped data to a new CSV file
output_path = "/Users/praveen/Downloads/grouped_by_case_name_with_parsed_vectors.csv"
grouped_data.to_csv(output_path, index=False)
print(f"Grouped data with parsed vector IDs saved to {output_path}.")
```
